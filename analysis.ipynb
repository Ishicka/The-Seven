# Unemployment vs Delinquency regression

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Load datasets
unemployment_data = pd.read_csv('UNRATE.csv')
delinquency_data = pd.read_csv('DRSFRMACBS.csv')

# Convert 'DATE' to datetime
unemployment_data['DATE'] = pd.to_datetime(unemployment_data['DATE'])
delinquency_data['DATE'] = pd.to_datetime(delinquency_data['DATE'])

# Merge the datasets on 'DATE'
merged_data = pd.merge(unemployment_data, delinquency_data, on='DATE', how='inner')
```

```python
# Extracting data for regression
X = merged_data['UNRATE'].values  # Unemployment Rate
y = merged_data['DRSFRMACBS'].values  # Mortgage Delinquency Rate

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(X, y)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, intercept + slope * X, color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Mortgage Delinquency Rate (%)')
plt.title('Regression Analysis: Unemployment Rate vs Mortgage Delinquency Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Results
results = {
    "Slope": slope,
    "Intercept": intercept,
    "R-squared": r_value**2,
    "P-value": p_value,
    "Standard Error": std_err
}
results
```
